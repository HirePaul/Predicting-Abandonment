```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
one_df = pd.read_excel('project_680..999.xlsx')
two_df = pd.read_excel('project_1000-4999.xlsx')
three_df = pd.read_excel('project_5000Up.xlsx')
#project_list = one_df['Project'].tolist() + two_df['Project'].tolist() + three_df['Project'].tolist()
size_list = one_df['Size'].tolist() + two_df['Size'].tolist() + three_df['Size'].tolist()
size_list = sorted(size_list)
```

```python
#for i in range(len(size_list)):
    #size_list[i] = size_list[i]/1000
# np.histogram(sizes)
plt.hist(size_list[:int(len(size_list)/2)], bins=100)
plt.show()
```

```python
print("Range for Plot above:",size_list[0],"-",size_list[int(len(size_list)/2)-1])
print("Number of data points above:", len(size_list[:int(len(size_list)/2)]))
```

```python
plt.hist(size_list[int(len(size_list)/2):-10000], bins=100)
plt.show()
```

```python
print("Range for Plot above:",size_list[int(len(size_list)/2)],"-",size_list[-10001])
print("Number of data points above:", len(size_list[int(len(size_list)/2):-10000]))
```
